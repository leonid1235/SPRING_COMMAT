## I. Метод Эйлера для ОДУ первого порядка.

Рассмотрим уравнение первого порядка

$$
\frac{d u}{d t} = \lambda u
$$

С начальным условием $u(t=0) = u_0$.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def euler_solve(lam, u0, T, dt):
    """Решает $du/dt = \lambda u$ на $0 < t < T$ с $u(t=0) = u0$ при помощи явного метода Эйлера."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y = np.empty(num_steps+1)
    y[0] = u0
    for k in range(num_steps):
        y[k+1] = y[k] + dt*lam*y[k]
    return tt, y
```

```python
lam = -0.5
tt, y = euler_solve(lam, u0=1.0, T=5, dt=0.1/abs(lam))
plt.plot(tt, y, 'o--', label='numeric solution')
plt.plot(tt, np.exp(lam*tt), '-', lw=2, label='ground truth')
plt.legend(loc='best')
plt.grid(True)
```

# Решение задания 1

Начнем с решения уравнения с использованием неявной схемы Эйлера. В неявной схеме значение $ u $ на следующем временном шаге определяется через значение на текущем временном шаге. Уравнение для неявной схемы выглядит следующим образом:

$
u_{n+1} = u_n + \Delta t \cdot \lambda \cdot u_{n+1}
$

где $ u_n $ - значение $ u $ на текущем временном шаге, $ u_{n+1} $ - значение $ u $ на следующем временном шаге, $ \Delta t $ - шаг по времени, а $ \lambda $ - коэффициент в уравнении.

Для решения этого уравнения для $ u_{n+1} $ мы можем использовать итерационный метод. Одним из самых простых методов является метод простой итерации (метод последовательных приближений), который выглядит так:

$
u_{n+1} = u_n + \Delta t \cdot \lambda \cdot u_{n+1}^*
$

где $ u_{n+1}^* $ - начальное приближение для $ u_{n+1} $.

Реализуем это в коде:


```python
import numpy as np
import matplotlib.pyplot as plt

def implicit_euler_solve(lam, u0, T, dt):
    """Решает du/dt = lambda u на 0 < t < T с u(t=0) = u0 при помощи неявного метода Эйлера."""
    num_steps = int(T / dt)  # Количество временных шагов
    tt = np.linspace(0.0, T, num_steps+1)  # Временные точки
    y = np.zeros(num_steps+1)  # Массив для значений u
    
    # Начальное значение
    y[0] = u0
    
    # Итерационное решение
    for i in range(1, num_steps+1):
        # Начальное приближение для неявной схемы - предыдущее значение
        y_guess = y[i-1]
        # Итерационное уточнение
        y_new = y[i-1] / (1 - lam * dt)  # Метод простой итерации
        # Обновляем значение u
        y[i] = y_new
    
    return tt, y

# Теперь нарисуем графики с разными значениями шага, и отдельно при lambda*tau > 2.

# Заданные параметры
lam = -0.5
T = 8/abs(lam)
u0 = 1.0

# Значения шага интегрирования
dt_values = [0.5/abs(lam), 1.0/abs(lam), 2.1/abs(lam)]

# Нарисуем графики
plt.figure(figsize=(10, 6))
for dt in dt_values:
    tt, y = implicit_euler_solve(lam, u0, T, dt)
    plt.plot(tt, y, 'o--', label=f'dt={dt}')
plt.plot(tt, np.exp(lam*tt), '-', lw=2, label='ground truth')
plt.title("Решение уравнения du/dt = lambda*u с неявной схемой Эйлера")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.legend()
plt.grid(True)
plt.show()
```

Обсудим поведение решения при различных значениях шага и при $ \lambda \cdot \Delta t > 2 $.

- При маленьких значениях шага (например, $ \Delta t = 0.5/ \lambda $) решение должно быть достаточно точным, так как мы используем маленькие шаги и метод простой итерации должен сходиться к точному значению.
- При больших значениях шага (например, $ \Delta t = 2.1/ \lambda $) решение может быть менее точным из-за того, что мы используем приближение метода простой итерации.
- Когда $ \lambda \cdot \Delta t > 2 $, неявная схема Эйлера становится неустойчивой, и решение начинает экспоненциально возрастать.

Добавим эту проверку в код и нарисуем график для случая $ \lambda \cdot \Delta t > 2 $:


```python
# Теперь нарисуем график для случая lambda*dt > 2
lam = -0.5 
dt = 2.1/abs(lam)  # Значение шага, при котором lambda*dt > 2
T = 8/abs(lam)
tt, y = implicit_euler_solve(lam, u0, T, dt)

plt.figure(figsize=(8, 6))
plt.plot(tt, y, 'o--', label='numeric solution')
plt.plot(tt, np.exp(lam*tt), '-', lw=2, label='ground truth')
plt.title("Решение уравнения du/dt = lambda*u с неявной схемой Эйлера (lambda*dt > 2)")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.legend()
plt.grid(True)
plt.show()
```

# Решение задания 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

def euler_solve2(a, u0, T, dt):
    """Solve the system du/dt = Au via an explicit Euler scheme.
    
    Parameters
    ----------
    a : ndarray, shape(ndim, ndim)
        The matrix of the l.h.s.
    u0 : ndarray, shape(ndim,)
        Initial condition
    T : float
        construct the solution for $t\in [0, T)$
    dt : float
        Integration step size $\tau$
        
    Returns
    -------
    t : ndarray, shape (n,)
        Integration times
    y : ndarray, shape (n, ndim)
        Solution of the FD system. 
        y[k, :] is the solution at t[k].
    """
    a = np.asarray(a, dtype=float)
    u0 = np.asarray(u0, dtype=float)
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    ndim = a.shape[0]
    y = np.zeros((num_steps+1, ndim))
    y[0] = u0
    for k in range(num_steps):
        y[k+1] = y[k] + dt * np.dot(a, y[k])
    return tt, y

def mat_exp_solve(a, u0, tt):
    """Construct the solution of $du/dt = A u$ with $u(t=0) = u_0$ at times `tt`.
    
    Parameters
    ----------
    a : ndarray, shape (ndim, ndim)
    
    u0 : ndarray, shape (ndim,)
    
    tt : ndarray, shape (n,)
        The values of $t$
        
    Return
    ------
    u : ndarray, shape (n, ndim)
        u[:, k] is $\exp(t[k] A) u_0$
    """
    u = np.zeros((len(tt), len(u0)))
    for i, t in enumerate(tt):
        u[i] = np.dot(expm(a * t), u0)
    return u


# Solve via Euler's method, compare to the matrix exponential
a = np.array([[-1, 1],
              [0, -1]], dtype=float)
t, y  = euler_solve2(a, u0=[1, 1], T=3, dt=0.1)
ym = mat_exp_solve(a, [1, 1], t)

plt.plot(t, y[:, 0], 'o', label='0, Euler')
plt.plot(t, ym[:, 0], '-', label='0, expm')

plt.plot(t, y[:, 1], 's', label='1, Euler')
plt.plot(t, ym[:, 1], '-', label='1, expm')

plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Код решает систему линейных дифференциальных уравнений первого порядка с матрицей $ A $, не зависящей от времени, с использованием явной схемы Эйлера в функции `euler_solve2`. Функция `mat_exp_solve` возвращает решение через матричную экспоненту. Результаты метода Эйлера и метода матричной экспоненты сравниваются для системы с определенной матрицей $ A $.


## III. Жесткие системы

Рассмотрим линейную систему, $du/dt = Au$, с матрицей правой части 
$$
A = \begin{bmatrix} -10 & 10 \\ 32 & -499 \end{bmatrix}
$$

с начальным условием $u = (1, 0)^T$.

Система называется жесткой, если для всех $(x, \vec{y}(x))$ (на решениях системы) собственные значения матрицы Якоби системы  $\mathbf{J}(x)$ (в данном случае она равна $A$)  удовлетворяют условиям

 $\max _j\left|\operatorname{Re} \lambda_j\right| / \min _k\left|\operatorname{Re} \lambda_k\right| \gg 1, \quad \operatorname{Re} \lambda_j<0$, $\max _j\left|\operatorname{Im} \lambda_j\right| \ll \max _k\left|\operatorname{Re} \lambda_k\right| \forall j, k$. 
 
 Число $s=\max _i\left|\operatorname{Re} \lambda_i\right| / \min _k\left|\operatorname{Re} \lambda_k\right|$ называется числом жесткости системы.

Пример.

Для решения задачи Коши системы ОДУ используется численный метод Рунге-Кутты, заданный таблицей Бутчера:
$$
\left.\begin{array}{c}
u^{\prime}=-800 u+4 \cdot 10^{-2} v+2 \cdot 10^{-2} w, u(0)=0, \\
v^{\prime}=-5 v-3 w, v(0)=4, \\
w^{\prime}=v-w, w(0)=6,
\end{array}\right\} \quad \quad \begin{array}{l|ll}
1 / 5 & 1 / 5 & 0 \\
4 / 5 & 3 / 5 & 1 / 5 \\
\hline & 1 / 2 & 1 / 2
\end{array}
$$
Получим для него функцию и условие устойчивости. Вычислим число жёсткости.

$$
R(z)=\frac{1+\frac{3}{5} z+\left(\frac{1}{2}-\frac{2}{5}+\frac{1}{25}\right) z^2}{1-\frac{2}{5} z+\frac{1}{25} z^2}
$$
Матрица $\mathbf{J}(x)$ (см. формулу (7.1.3)) постоянна и ее собственные значения $\lambda_1=-2, \lambda_2=-4, \lambda_3=-800$ лежат на действительной оси. Так как $z_i=\lambda_i h$, то и функцию устойчивости достаточно исследовать на действительной оси. 

Область устойчивости определяется из условия $|R(z)| \leq 1$. Получаем

 $\operatorname{Im} z_i=0, \operatorname{Re} z_i\left(1+\operatorname{Re} z_i / 10\right) \leq 0, i=1,2,3$.


Шаг интегрирования, который удовлетворяет всем условиям $h \in(0,10 / 800]$. 

Число жесткости $s=800 / 2=400$.

# Решение задания 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для решения системы методом Эйлера с фиксированным шагом
def euler_solve2(A, u0, tau, T):
    N = int(T / tau) + 1
    t = np.linspace(0, T, N)
    u = np.zeros((2, N))
    u[:, 0] = u0
    for i in range(1, N):
        u[:, i] = u[:, i-1] + tau * np.dot(A, u[:, i-1])
    return t, u

# Функция для решения системы матричным экспонентом
def mat_exp_solve(A, u0, tau, T):
    t = np.arange(0, T, tau)
    u = np.zeros((2, len(t)))
    u[:, 0] = u0
    for i in range(1, len(t)):
        u[:, i] = np.dot(np.linalg.expm(A * t[i]), u0)
    return t, u

# Матрица A
A = np.array([[0, 1], [-1000, -1]])

# Найдем собственные значения матрицы A
eigenvalues = np.linalg.eigvals(A)
print("Собственные значения матрицы A:", eigenvalues)

# Проверим, является ли система жесткой
if np.max(np.abs(eigenvalues.real)) > 0:
    print("Система является жесткой.")
else:
    print("Система не является жесткой.")

# Решение системы с фиксированным шагом методом Эйлера
tau1 = 4e-3
tau2 = 4.5e-3
T = 1
u0 = np.array([1, 0])

t1, u1 = euler_solve2(A, u0, tau1, T)
t2, u2 = euler_solve2(A, u0, tau2, T)

# Построим графики
plt.figure(figsize=(12, 6))
plt.plot(t1, u1[0], label=r"$\tau = 4\cdot 10^{-3}$")
plt.plot(t2, u2[0], label=r"$\tau = 4.5\cdot 10^{-3}$")
plt.xlabel('t')
plt.ylabel('u')
plt.title('Решение системы методом Эйлера')
plt.legend()
plt.grid(True)
plt.show()
```

Теперь добавим реализацию неявной схемы Эйлера:

```python
# Реализация неявной схемы Эйлера
def implicit_euler_solve(A, u0, tau, T):
    N = int(T / tau) + 1
    t = np.linspace(0, T, N)
    u = np.zeros((2, N))
    u[:, 0] = u0
    I = np.eye(2)
    for i in range(1, N):
        u[:, i] = np.linalg.solve(I - tau * A, u[:, i-1])
    return t, u

# Решение неявной схемой Эйлера
tau = 4e-3
T = 1
u0 = np.array([1, 0])

t_imp, u_imp = implicit_euler_solve(A, u0, tau, T)

# Сравним результаты явной и неявной схемы
plt.figure(figsize=(12, 6))
plt.plot(t1, u1[0], label=r"Явная Эйлер: $\tau = 4\cdot 10^{-3}$")
plt.plot(t_imp, u_imp[0], label=r"Неявная Эйлер: $\tau = 4\cdot 10^{-3}$")
plt.xlabel('t')
plt.ylabel('u')
plt.title('Сравнение явной и неявной схем Эйлера')
plt.legend()
plt.grid(True)
plt.show()
```

##### Определение функции устойчивости и области устойчивости метода Рунге-Кутты

Для определения функции устойчивости метода Рунге-Кутты, заданного таблицей Бутчера:

\begin{array}{l|ll}
0 & 0 & 0 \\
2 & 3 / 2 & 1 / 2 \\
\hline & 3 / 4 & 1 / 4

$
R(z) = 1 + z \cdot (a_{21} + a_{22}z)
$

где $a_{21}$ и $a_{22}$ - элементы матрицы $A$.

##### Область устойчивости

- **A-устойчивость:** Для A-устойчивости нужно, чтобы для всех \(z\) находящихся в левой полуплоскости выполнялось \(|R(z)| < 1\).
- **L-устойчивость:** Для L-устойчивости нужно, чтобы для всех \(z\) находящихся на мнимой оси выполнялось \(|R(z)| \leq 1\).



```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta_stability_region(a):
    def R(z):
        return 1 + z * (a[1][0] + a[1][1] * z)
    
    # Plot the stability region
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    X, Y = np.meshgrid(x, y)
    Z = R(X + 1j * Y)

    plt.figure(figsize=(8, 8))
    plt.contour(X, Y, np.abs(Z), levels=[1], colors='b')
    plt.xlabel('Re')
    plt.ylabel('Im')
    plt.title('Stability Region')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

def check_A_stability(a):
    # A-stability check for z in left half-plane
    z_values = np.linspace(-10, 0, 1000)
    R_values = np.abs(1 + z_values * (a[1][0] + a[1][1] * z_values))
    max_R = np.max(R_values)
    if max_R < 1:
        print("The method is A-stable.")
    else:
        print("The method is not A-stable.")

def check_L_stability(a):
    # L-stability check for z on the imaginary axis
    z_values = np.linspace(-10, 10, 1000)
    R_values = np.abs(1 + z_values * (a[1][0] + a[1][1] * z_values))
    max_R = np.max(R_values)
    if max_R <= 1:
        print("The method is L-stable.")
    else:
        print("The method is not L-stable.")

# Define the Butcher table
a = np.array([[0, 0],
              [3/2, 1/2]])

runge_kutta_stability_region(a)
check_A_stability(a)
check_L_stability(a)
```

# Решение задания 5

Для начала, давайте реализуем метод Рунге-Кутты заданный таблицей Бутчера:

```python
import numpy as np

def runge_kutta_butcher(f, y0, t0, tf, h, butcher_table):
    """
    Реализация метода Рунге-Кутты с использованием таблицы Бутчера.
    
    Параметры:
    f : функция
        Функция, задающая систему ОДУ. Должна принимать аргументы (t, y), где t - текущее время, y - текущее значение.
    y0 : array_like
        Начальные условия.
    t0 : float
        Начальное время.
    tf : float
        Конечное время.
    h : float
        Шаг интегрирования.
    butcher_table : array_like
        Таблица Бутчера.
        
    Возвращает:
    t_values : array_like
        Массив значений времени.
    y_values : array_like
        Массив значений решения в соответствующие моменты времени.
    """
    a = butcher_table[:-1, :-1]
    b = butcher_table[-1, :-1]
    c = butcher_table[-1, -1]
    
    num_steps = int((tf - t0) / h)
    t_values = np.linspace(t0, tf, num_steps + 1)
    y_values = np.zeros((len(y0), num_steps + 1))
    y_values[:, 0] = y0
    
    for i in range(num_steps):
        t_i = t_values[i]
        y_i = y_values[:, i]
        k = np.zeros((len(y0), len(b)))
        
        for j in range(len(b)):
            k[:, j] = f(t_i + c * h, y_i + h * np.sum(a[j, :] * k, axis=1))
            
        y_values[:, i+1] = y_i + h * np.sum(b * k, axis=1)
    
    return t_values, y_values
```

Теперь определим функцию для системы ОДУ из условия:

```python
def system_ode(t, y):
    """
    Функция, задающая систему ОДУ из условия.
    
    Параметры:
    t : float
        Текущее время.
    y : array_like
        Текущее значение функций u, v, w.
        
    Возвращает:
    dydt : array_like
        Значения производных функций u, v, w в заданный момент времени t.
    """
    u, v, w = y
    du_dt = -400 * u
    dv_dt = 3e-2 * u - 10 * v + w
    dw_dt = -4e-2 * u - 12 * v - 2 * w
    dydt = np.array([du_dt, dv_dt, dw_dt])
    return dydt
```

```python
def stiffness_number(A, h):
    """
    Функция для вычисления числа жёсткости системы ОДУ.
    
    Параметры:
    A : array_like
        Матрица системы ОДУ.
    h : float
        Шаг интегрирования.
        
    Возвращает:
    stiffness : float
        Число жёсткости.
    """
    eigenvalues, _ = np.linalg.eig(A)
    lambda_max = np.max(np.abs(eigenvalues))
    stiffness = lambda_max * h / 2
    return stiffness

# Определим матрицу A
A = np.array([[-400, 0, 0],
              [3e-2, -10, 1],
              [-4e-2, -12, -2]])

# Зададим шаг интегрирования
h = 0.01  # Примерное значение шага, можно подобрать более точное значение

# Вычислим число жёсткости
stiffness = stiffness_number(A, h)
print("Число жёсткости:", stiffness)
```

# Решение задания 6

Для начала, мы преобразуем уравнение второго порядка в систему ОДУ первого порядка. Пусть $ u = x_1 $ и $ v = x_2 = \frac{{du}}{{dt}} $, тогда:

$
\begin{align*}
\frac{{dx_1}}{{dt}} &= x_2 \\
\frac{{dx_2}}{{dt}} &= -\omega^2 x_1
\end{align*}
$

Теперь мы можем написать код для численного решения данной системы уравнений, используя явную схему Эйлера.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
omega = 1.0  # Angular frequency
t_end = 10 * 2 * np.pi / omega  # End time
num_periods = 10  # Number of periods
num_steps = 1000  # Number of time steps
tau_values = [0.01, 0.005]  # Step sizes

# Function defining the system of ODEs
def f(t, y):
    return np.array([y[1], -omega**2 * y[0]])

# Euler's method
def euler_method(f, y0, t):
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    for i in range(1, len(t)):
        y[i] = y[i-1] + f(t[i-1], y[i-1]) * (t[i] - t[i-1])
    return y

# Initial conditions
y0 = np.array([1.0, 0.0])  # initial position and velocity

# Time array
t = np.linspace(0, t_end, num_steps)

# Plotting
plt.figure(figsize=(10, 6))
for tau in tau_values:
    # Euler's method
    y_euler = euler_method(f, y0, t)

    # Energy calculation
    energy = 0.5 * y_euler[:, 1]**2 + 0.5 * omega**2 * y_euler[:, 0]**2

    # Plot energy vs. time
    plt.plot(t, energy, label=f'Euler, step size = {tau}')

plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy vs. Time (Euler Method)')
plt.legend()
plt.grid(True)
plt.show()
```

Теперь мы реализуем метод Рунге-Кутта второго порядка и сравним его с методом Эйлера.

```python
# Runge-Kutta 2nd order method
def runge_kutta_2nd_order(f, y0, t):
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    for i in range(1, len(t)):
        dt = t[i] - t[i-1]
        k1 = f(t[i-1], y[i-1])
        k2 = f(t[i-1] + dt/2, y[i-1] + dt/2 * k1)
        y[i] = y[i-1] + dt * k2
    return y

plt.figure(figsize=(10, 6))
for tau in tau_values:
    # Runge-Kutta 2nd order method
    y_rk2 = runge_kutta_2nd_order(f, y0, t)

    # Energy calculation
    energy_rk2 = 0.5 * y_rk2[:, 1]**2 + 0.5 * omega**2 * y_rk2[:, 0]**2

    # Plot energy vs. time
    plt.plot(t, energy_rk2, label=f'RK2, step size = {tau}')

plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy vs. Time (Runge-Kutta 2nd Order Method)')
plt.legend()
plt.grid(True)
plt.show()
```

Теперь мы рассмотрим предиктор-корректорный метод. Для этого сначала определим предиктор, а затем выполним коррекцию в соответствии с законом сохранения энергии.

```python
# Predictor-Corrector method
def predictor_corrector(f, y0, t):
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    for i in range(1, len(t)):
        dt = t[i] - t[i-1]
        # Predictor (Euler's method)
        y_pred = y[i-1] + dt * f(t[i-1], y[i-1])
        # Corrector (energy conservation)
        y[i] = y[i-1] + dt * f(t[i], y_pred)
    return y

plt.figure(figsize=(10, 6))
for tau in tau_values:
    # Predictor-Corrector method
    y_pc = predictor_corrector(f, y0, t)

    # Energy calculation
    energy_pc = 0.5 * y_pc[:, 1]**2 + 0.5 * omega**2 * y_pc[:, 0]**2

    # Plot energy vs. time
    plt.plot(t, energy_pc, label=f'Predictor-Corrector, step size = {tau}')

plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy vs. Time (Predictor-Corrector Method)')
plt.legend()
plt.grid(True)
plt.show()
```

# Решение задания 8

```python
import numpy as np
import matplotlib.pyplot as plt


# Определим функции для численного решения и аналитического решения с использованием ряда Тейлора:

def numerical_solve_poisson(n, dx, phi_0, phi_prime_0):
    x = np.arange(0, 10 + dx, dx)
    phi = np.zeros_like(x)
    phi_prime = np.zeros_like(x)

    phi[0] = phi_0
    phi_prime[0] = phi_prime_0

    for i in range(1, len(x)):
        phi[i] = phi[i - 1] + phi_prime[i - 1] * dx
        phi_prime[i] = phi_prime[i - 1] + (np.exp(phi[i - 1]) - n[i - 1]) * dx

    return x, phi

def analytical_solve_poisson(n, dx, phi_0, phi_prime_0):
    x = np.arange(0, 10 + dx, dx)
    phi_analytical = np.zeros_like(x)
    phi_analytical[0] = phi_0
    phi_prime_analytical = np.zeros_like(x)
    phi_prime_analytical[0] = phi_prime_0

    for i in range(1, len(x)):
        phi_analytical[i] = phi_analytical[i - 1] + phi_prime_analytical[i - 1] * dx
        phi_prime_analytical[i] = phi_prime_analytical[i - 1] + (1 + np.exp(-3 * (x[i] - 5) ** 2) - np.exp(phi_analytical[i - 1])) * dx

    return x, phi_analytical


# Теперь можно использовать эти функции для численного и аналитического решений и сравнить результаты:

# Параметры
dx = 0.01
phi_0 = 0
phi_prime_0 = 0

# Создание сетки
x = np.arange(0, 10 + dx, dx)
n = 1 + np.exp(-3 * (x - 5) ** 2)

# Численное решение
x_num, phi_num = numerical_solve_poisson(n, dx, phi_0, phi_prime_0)

# Аналитическое решение с помощью ряда Тейлора
x_analytical, phi_analytical = analytical_solve_poisson(n, dx, phi_0, phi_prime_0)

# Визуализация результатов
plt.figure(figsize=(10, 6))
plt.plot(x_num, phi_num, label='Numerical Solution', linestyle='-', color='b')
plt.plot(x_analytical, phi_analytical, label='Analytical Solution (Taylor Series)', linestyle='--', color='r')
plt.xlabel('x')
plt.ylabel('Phi(x)')
plt.title('Comparison of Numerical and Analytical Solutions')
plt.legend()
plt.grid(True)
plt.show()
```

Этот код вычисляет численное и аналитическое решения уравнения Пуассона и построит их графики для сравнения.

## Задание 11. Закон естественного радиоактивного распада.

Промоделируйте радиоактивный распад группы атомов. Постройте графики количества оставшихся атомов с течением для случаев $N_0$ = 10, 100, 1000, 1000000. Постройте их также в логарифмическом масштабе по N.

Эксперименты с тяжелыми атомами говорят, что их ядра распадаются, выбрасывая частицы и превращаясь в более легкие. Распад ядра происходит спонтанно и практически не зависит от внешних условий, в которых оно находится. Число распадов, регистрируемых в единицу времени, не зависит от внешних полей, концентрации радиоактивных атомов, химической структуры молекул, в которые входят радиоактивные атомы, а зависит только от числа атомов, ядра которых способны распадаться. Вероятность естественного распада ядра определяется его устройством и не зависит от того, что происходит вне ядра. Эта вероятность может быть описана некоторым числом, характерным для данных ядер. Таково главное свойство естественного радиоактивного распада. 

Вопросы для численного эксперимента на математической модели можно поставить следующим образом:

1. Как меняется со временем число нераспавшихся ядер, если в начальный момент их было $N_{0}$ ?
2. Каково среднее время жизни $T_{1}$ ядра?
3. Через какое время $T_{2}$ распадется половина ядер?
Ответы на первый и последний вопросы достаточно легко проверить в натурном эксперименте с быстро распадающимися ядрами.


Приступим к построению математической модели. В принятой концепции явления можно сказать, что на малом интервале времени $\Delta t$ число распавшихся ядер $-\Delta N=N(t)-N(t+\Delta t)$ пропорционально $N^{*} \Delta t .$
Кроме того, это число пропорционально числу нераспавшихся ядер в данный момент времени, поскольку каждое ядро независимо от других может испытать распад на рассматриваемом интервале времени Dt. 

Учитывая сказанное, для достаточно малого интервала времени можно написать следующее уравнение радиоактивного распада  

$\Delta N=\frac{N}{\tau} \cdot \Delta t$

Коэффициент пропорциональности записан в таком виде, чтобы константа вещества t имела размерность времени. Она определяет, насколько быстро распадаются ядра данного вещества. В теории радиоактивного распада обычно используют другую постоянную $\lambda=1 / \tau$, называемую радиоактивной постоянной.

Нам кажется использование $\tau$ более удобным, тем более что оно, как покажет эксперимент, имеет простой физический смысл. Если $\tau-$ велико, то за время $\Delta t$ распадется мало ядер, и наоборот, если $\tau$ - мало, то ядра распадаются быстро. Во втором случае требуется выбирать достаточно маленькое $\Delta t$, чтобы $\Delta N$ было малым и уравнение справедливым. Величина $\tau$ должна быть измерена в натурном эксперименте по подсчету числа распадов в заданном образце на небольшом интервале времени и вычислена по формуле.  

Опыт дает очень разные значения этой константы для разных веществ. Так, для радона $\tau=5,56$ дня, для радия - 2640 лет, для урана - 7,5\bullet10 $^{9}$ лет. Для полного определения функции $N(t)$ нужно задать число ядер вещества в начальный момент:

$$
t=0: N=N_{0}
$$

Уравнения определяют математическую модель радиоактивного распада. Теперь в модели нужно перейти к относительным единицам, чтобы уменьшить число параметров задачи и представить результаты численного эксперимента в относительных единицах.
В качестве характерного времени процесса можно взять $\tau$ - оно будет служить новой единицей времени: размерное время $t$, измеряемое в секундах, представим в виде $\mathrm{t}=\tau \cdot \tilde{t}, \tilde{t}-6$ езразмерное время.

В качестве характерного числа атомов возьмем их количество в момент $t=0$, т.е. представим число атомов в произвольный момент времени в виде $N=N_{0}$* $\widetilde{N} .$ Число $\widetilde{N}$ показывает долю оставшихся ядер от их числа в начальный момент. Подставим размерные величины:

$$
\Delta \widetilde{N}=-\widetilde{N} \cdot \Delta t, \tilde{t}=0: \tilde{N}=1
$$

Все входящие в уравнение величины безразмерные, поэтому в дальнейшем знак безразмерной величины “ " писать не будем.

Перейдем к описанию алгоритма вычисления числа ядер, не распавшихся к моменту времени $t$. Вдоль оси времени будем продвигаться малыми, но конечными шагами величиной $\Delta t$, обозначая соответствующие моменты времени $t_{n}=n \Delta t$ и число ядер $N_{n} .$ Связь между этими числами задается уравнением:

$$
\Delta N=N_{n+1}-N_{n}=-N_{n} \cdot \Delta t
$$

Запись в правой части числа ядер на левом конце временного интервала $\Delta t$ может привести к заметной ошибке, если на этом интервале происходит заметное изменение числа ядер.

$$
N_{0}=1, N_{n}+1=N_{n}-N_{n} \cdot \Delta t, n=1,2, \ldots
$$

Система разностных уравнений определяет число нераспавшихся ядер в последовательные моменты времени. Обратим внимание на то, что она не содержит характеристик вещества: полученные при численном эксперименте свойства распада будут универсальными.
Индивидуальность ядер проявится только через масштаб времени t.

Вернемся к вопросам, поставленным перед численным экспериментом. Для определения зависимости $N(t)$ по вычисленным значениям $N_{a}$ нужно построить плавную кривую процесса. Чтобы вычислить среднее время жизни ядра $T_{1}$, нужно просуммировать время жизни каждого из ядер и разделить его на полное число распавшихся ядер. Время жизни $t_{n}$ имеют те ядра, которые распались на интервале $t_{n+1}-t_{n} .$ Их число есть $N_{n}-N_{n+1}  .$ Отсюда получим

$$
T_{1}=\left(\sum_{n=0}^{\infty} t_{n}\left(N_{n}-N_{n+1}\right)/N_{0}\right)=(\Delta t)^{2} \sum_{n=0}^{\infty} n N_{n}
$$

Для вычисления времени $T_{2}$, за которое распадается половина начального числа ядер, нужно найти для определяемой в численном эксперименте функции $N(t)$ решение уравнения $N(t)=0,5 .$

## Задание 12. Моделирование броуновского движения.



### Постановка задачи.

Броуновским движением называют наблюдающееся под микроскопом непрерывное хаотическое движение мелких частиц, взвешенных в жидкости или газе. Молекулы среды находятся в непрерывном тепловом движении и при столкновении с частицей передают ей соответствующий импульс. Если размеры частицы достаточно малы (меньше $10^{-4}$ см), то удары, получаемые частицей с разных сторон, не компенсируют друг друга в силу их небольшого количества. Под влиянием этих флуктуаций давления на поверхности частицы она приходит в беспорядочное движение, направление и скорость которого меняются с большой частотой ($10^{12} c^{-1}$). 

Чтобы анализировать это явление, экспериментатор фиксирует положение частицы через равные промежутки времени, например, через 30 сек., и соединяет их прямыми линиями. 

В каждый момент времени частица может с одинаковой вероятностью переместиться по любому направлению. Величина этого смещения также случайна и зависит от свойств жидкости, ее температуры и размеров частицы. Для фиксированной системы имеется некоторое среднее за один такт наблюдения смещение l, которое возрастает с ростом T.   

### Численная модель.

Положение броуновской частицы будем задавать в декартовой системе координат. В начальный момент пусть частица находится в начале координат. Смещение частицы вдоль каждой из осей является случайным со средним по величине смещением / и равновероятным в обоих направлениях. Это смещение будем формировать с помощью датчика случайных чисел с равномерным распределением от 0 до $1 .$

$$
d x=2 \cdot l \cdot(\text { random }-0.5), \quad d y=2 \cdot l \cdot(\text { random }-0.5) ;
$$

Координаты нового положения частицы в последовательные моменты времени будем определять так: $x_{n+1}=x_{n}+d x, y_{n+1}=y_{n}+d y .$
Соответствующий момент времени будет $t_{n}=\Delta t \cdot n$. 

Таким образом, можно сказать, что время пропорционально числу шагов по фиксации положения частицы. Расстояние $L$, на которое удалилась частица от начала координат, очевидно, равно $\sqrt{x_{n}^{2}+y_{n}^{2}} .$ За интервал времени $\mathrm{T}=\mathrm{t}_{n}-\mathrm{t}_{\mathrm{m}}$ квадрат смещения броуновской частицы равен $\mathrm{S} .$

В численном эксперименте нужно исследовать зависимость $S$ от $\mathrm{T} .$  Ниже приведён пример моделирования броуновского движения на плоскости.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
steps = 5000
random.seed(42)
x,y = np.cumsum(np.random.randn(steps)), np.cumsum(np.random.randn(steps))
points = 10
ip = lambda x, steps, points: np.interp(np.arange(steps*points), 
                                        np.arange(steps)*points, 
                                        x)
X, Y = ip(x, steps, points), ip(y, steps, points)
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.set_title('Brownian Motion')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(X, Y, color='blue',
        marker='o',  markersize=1)  
```
